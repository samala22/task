```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
wq = pd.read_csv('wine.csv')
```

```python
wq['quality'] = wq['quality'].replace({"good": 1, "bad": 0})
```

```python
print(wq.head())
print(wq.info())
print(wq.describe())
print(wq.isnull().sum())
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

sns.barplot(x='quality', y='fixed acidity', data=wq, ax=axes[0, 0])
sns.barplot(x='quality', y='volatile acidity', data=wq, ax=axes[0, 1])
sns.barplot(x='quality', y='citric acid', data=wq, ax=axes[0, 2])
sns.barplot(x='quality', y='residual sugar', data=wq, ax=axes[1, 0])
sns.barplot(x='quality', y='chlorides', data=wq, ax=axes[1, 1])
sns.barplot(x='quality', y='free sulfur dioxide', data=wq, ax=axes[1, 2])
```

```python
plt.tight_layout()  # Adjust subplot layout
plt.show()
```

```python
sns.countplot(data=wq, x='quality')
plt.show()
```

```python
X = wq.drop("quality", axis=1)
y = wq["quality"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
```

```python
model = RandomForestClassifier(criterion="gini", n_estimators=90, random_state=0)
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print("Model Accuracy:", accuracy)
```
